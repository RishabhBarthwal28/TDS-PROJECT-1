**Q10**

```python
from sklearn.linear_model import LinearRegression

# Prepare the data for regression
X = df['public_repos'].values.reshape(-1, 1)  # Independent variable: public repositories
y = df['followers'].values  # Dependent variable: followers

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# The coefficient represents the estimated increase in followers per additional repository
followers_per_repo = model.coef_[0]
followers_per_repo
```

**Q11**

```python
# Convert 'has_projects' and 'has_wiki' columns to numerical (1 for 'true', 0 for 'false')
repdf['has_projects'] = repdf['has_projects'].apply(lambda x: 1 if x == 'true' else 0)
repdf['has_wiki'] = repdf['has_wiki'].apply(lambda x: 1 if x == 'true' else 0)

# Calculate the correlation between 'has_projects' and 'has_wiki'
projects_wiki_correlation = repdf['has_projects'].corr(repdf['has_wiki'])

# Round to 3 decimal places
projects_wiki_correlation_rounded = round(projects_wiki_correlation, 3)
projects_wiki_correlation_rounded
```

**Q12**

```python
# Convert 'hireable' column to boolean if it's in string format ('true'/'false')
df['hireable'] = df['hireable'].apply(lambda x: True if x == 'true' else False)

# Calculate the average 'following' for hireable and non-hireable users
avg_following_hireable = df[df['hireable'] == True]['following'].mean()
avg_following_non_hireable = df[df['hireable'] == False]['following'].mean()

# Calculate the difference between the two averages, rounded to 3 decimal places
following_difference = round(avg_following_hireable - avg_following_non_hireable, 3)
following_difference
```

**Q13**

```python
from sklearn.linear_model import LinearRegression

# Filter out users without bios and calculate the word count of each bio
df_with_bio = df[df['bio'].notna() & (df['bio'] != "")].copy()
df_with_bio['bio_word_count'] = df_with_bio['bio'].apply(lambda x: len(x.split()))

# Prepare data for regression
X_bio_words = df_with_bio['bio_word_count'].values.reshape(-1, 1)  # Independent variable: bio word count
y_followers = df_with_bio['followers'].values  # Dependent variable: followers

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X_bio_words, y_followers)

# Get the slope (impact) of bio word count on followers
bio_word_slope = round(model.coef_[0], 3)
bio_word_slope
```

**Q16**

```python
# Filter out users without a name and trim whitespace
df_with_names = df[df['name'].notna() & (df['name'] != "")].copy()
df_with_names['surname'] = df_with_names['name'].apply(lambda x: x.strip().split()[-1])

# Count occurrences of each surname and find the most common one(s)
surname_counts = df_with_names['surname'].value_counts()
max_count = surname_counts.max()
most_common_surnames = surname_counts[surname_counts == max_count].index.tolist()

# Sort the surnames alphabetically and join them as a comma-separated string
most_common_surnames_sorted = ', '.join(sorted(most_common_surnames))
most_common_surnames_sorted
```
